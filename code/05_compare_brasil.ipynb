## Importar os pacotes que serão usados

```python
import numpy as np
import pandas as pd
import glob
```

## Criar um dataframe contendo todas as classes presentes no mapa completo

```python
path = "../confusion_matrix/*.csv"
files = glob.glob(path)

classes1 = []
classes2 = []

for filename in files:
  base = pd.read_csv(files[0], index_col="Unnamed: 0")

  # calcula os valores de pixel unicos de cada mapa
  classes1.extend(base.columns)
  classes2.extend(base.index)

classes1 = np.unique(classes1)
classes2 = np.unique(classes2)

brasil_geral = pd.DataFrame(0, index=classes2, columns=classes1)
```

## Popular esse dataframe com as tabulações cruzadas apra cada bloco

```python
path = "../confusion_matrix/*.csv"
files = glob.glob(path)
for filename in files:
    brasil_bloco = pd.read_csv(files[0], index_col="Unnamed: 0")

    for column in brasil_bloco.columns:
        for index in brasil_bloco.index:
            brasil_geral.at[index, column] = brasil_geral.at[index, column] + brasil_bloco.at[index, column]

brasil_geral = brasil_geral.loc[(brasil_geral != 0).any(axis=1)]
brasil_geral = brasil_geral.loc[:, (brasil_geral != 0).any(axis=0)]
```
